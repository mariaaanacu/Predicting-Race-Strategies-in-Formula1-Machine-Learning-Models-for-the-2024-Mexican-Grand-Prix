```python
import os
import pandas as pd

def process_car_data(file_path, output_file_path):
    output_dir = os.path.dirname(output_file_path)
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    df = pd.read_excel(file_path)

    df['Year'] = df['Year'].astype(int)
    df['DriverName'] = df['DriverName'].astype(str)
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce') 
    df['RPM'] = df['RPM'].astype(int)
    df['Speed'] = df['Speed'].astype(float)
    df['nGear'] = df['nGear'].astype(int)
    df['Throttle'] = df['Throttle'].astype(float)
    df['Brake'] = df['Brake'].astype(bool)
    df['DRS'] = df['DRS'].astype(int)
    df['Source'] = df['Source'].astype(str)
    df['Time'] = pd.to_timedelta(df['Time'], errors='coerce')  
    df['SessionTime'] = pd.to_timedelta(df['SessionTime'], errors='coerce')  

    df.to_excel(output_file_path, index=False)
    print(f"Fișierul {output_file_path} a fost procesat și salvat.")

    print(f"\nTipurile de date pentru {output_file_path}:\n")
    print(df.dtypes)

file_paths = [
    "C:\\Users\\admin\\Desktop\\Formula 1 Prediction Challenge 2024 Mexico Grand Prix\\ocean_mexican_gp_data_challenge_datasets\\races_2024\\car_data_2024_processed_final.xlsx",
    "C:\\Users\\admin\\Desktop\\Formula 1 Prediction Challenge 2024 Mexico Grand Prix\\ocean_mexican_gp_data_challenge_datasets\\mexican_gp_race_datasets_2018-2019-2021-2022-2023\\car_data_mexico_grand_processed_final.xlsx",
    "C:\\Users\\admin\\Desktop\\Formula 1 Prediction Challenge 2024 Mexico Grand Prix\\ocean_mexican_gp_data_challenge_datasets\\mexican_gp_qualifying_datasets_2018-2019-2021-2022-2023\\car_data_mexico_grand_prix_qualifying_processed_final.xlsx"
]

for file_path in file_paths:
    process_car_data(file_path, file_path)
```

```python
import os
import pandas as pd

def process_position_data(file_path, output_file_path):
    output_dir = os.path.dirname(output_file_path)
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    df = pd.read_excel(file_path)

    df['Year'] = df['Year'].astype(int)

    timedelta_cols = ['Time', 'SessionTime']
    for col in timedelta_cols:
        df[col] = pd.to_timedelta(df[col], errors='coerce')
        df[col].fillna(pd.Timedelta(seconds=0), inplace=True)

    df['DriverName'] = df['DriverName'].astype(str)
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce') 
    df['Status'] = df['Status'].astype(str)
    df['X'] = df['X'].astype(float)
    df['Y'] = df['Y'].astype(float)
    df['Z'] = df['Z'].astype(float)
    df['Source'] = df['Source'].astype(str)

    df.to_excel(output_file_path, index=False)
    print(f"File {output_file_path} has been processed and saved.")

    print(f"\nData types for {output_file_path}:\n")
    print(df.dtypes)

file_paths = [
    "C:\\Users\\admin\\Desktop\\Formula 1 Prediction Challenge 2024 Mexico Grand Prix\\ocean_mexican_gp_data_challenge_datasets\\races_2024\\position_2024_processed_final.xlsx",
    "C:\\Users\\admin\\Desktop\\Formula 1 Prediction Challenge 2024 Mexico Grand Prix\\ocean_mexican_gp_data_challenge_datasets\\mexican_gp_race_datasets_2018-2019-2021-2022-2023\\position_mexico_grand_prix_2018-2019-2021-2022-2023_processed_final.xlsx",
    "C:\\Users\\admin\\Desktop\\Formula 1 Prediction Challenge 2024 Mexico Grand Prix\\ocean_mexican_gp_data_challenge_datasets\\mexican_gp_qualifying_datasets_2018-2019-2021-2022-2023\\position_mexico_grand_prix_2018-2019-2021-2022-2023_qualifying_processed_final.xlsx"
]

for file_path in file_paths:
    process_position_data(file_path, file_path)
```
